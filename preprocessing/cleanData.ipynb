```python
import pandas as pd
```

```python
df= pd.read_csv("allDataSets.csv")
```

```python
df.duplicated().sum()
df = df.drop_duplicates()
```

```python
df.isnull().sum()
#bmi                   4163
#smoking_status       13292
```

```python
# bring median of bmi 
median_bmi=df['bmi'].median()
# put this median for all missing values
df['bmi']=df['bmi'].fillna(median_bmi)
```

```python
# smoking_status is categorical variable ==> يعنى متغير بيقسم الداتا ل فئات 
# put Unknown in all missing values 
df['smoking_status']=df['smoking_status'].fillna("Unknown")
df.to_csv("allDataSets.csv", index=False)
```
